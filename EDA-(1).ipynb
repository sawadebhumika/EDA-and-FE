```python
data_copy = data_copy.drop(data_copy.index[10472])
```

```python
data_copy.shape
```

```python
data_copy["Reviews"].dtype
```

```python
data_copy["Reviews"] = data_copy["Reviews"].astype("int")  #convrting data-type
```

```python
data_copy["Reviews"].dtype
```

```python
data_copy["Size"].dtype
```

```python
data.head()
```

```python
data_copy["Size"].unique()
```

```python
data_copy["Size"] = data_copy["Size"].str.replace("M","000")
```

```python
data_copy["Size"] = data_copy["Size"].str.replace("k","")
```

```python
import numpy as np
data_copy["Size"] = data_copy["Size"].str.replace("Varies with device",str(np.nan))
```

```python
data_copy["Size"].unique()
```

```python
data_copy["Size"] = data_copy["Size"].astype("float")
```

```python
data_copy["Size"].dtype
```

```python
data_copy["Size"]*1000
```

```python
for i in data_copy["Size"]:
    if i < 10:
        data_copy["Size"] = data_copy["Size"].replace(i,i*1000)
```

```python
data_copy.columns
```

```python
data_copy["Installs"] = data_copy["Installs"].str.replace("+","")
```

```python
data_copy["Installs"].unique()
```

```python
data_copy["Price"].unique()
```

```python
char_to_removes = ["+",",","$"]
cols_to_clean = ["Installs","Price"]

for item in char_to_removes:
    for col in cols_to_clean:
        data_copy[col] = data_copy[col].str.replace(item,"")
```

```python
data_copy["Price"].unique()
```

```python
data_copy["Installs"] = data_copy["Installs"].astype("int")
data_copy["Price"] = data_copy["Price"].astype("float")
```

```python
data_copy.dtypes
```

```python
data_copy["Last Updated"].dtype
```

```python
data_copy["Last Updated"] = pd.to_datetime(data_copy["Last Updated"])
```

```python
data_copy["Last Updated"].dtype
```

```python
data_copy["Year"] = data_copy["Last Updated"].dt.year
```

```python
data_copy.head()
```

```python
data_copy["Android Ver"].unique()
```

```python
data_copy["Android Ver"] = data_copy["Android Ver"].str.replace("Varies with device","")
```

```python
char_to_replace = [" and up","Varies with device"]
for i in char_to_replace:
    data_copy["Android Ver"] = data_copy["Android Ver"].str.replace(i,"")
```

```python
data_copy["Android Ver"].unique()
```

```python
# What are the top  most installed in each popular category

dfa = df.groupby(["Category","App"])["Installs"].sum().reset_index()
```

```python
dfa=dfa.sort_values("Installs",ascending=False)
```

```python
dfa[dfa.Category=="GAME"][:5]
```

```python
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```
